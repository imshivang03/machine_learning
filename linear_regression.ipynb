```python
## Independent and Dependent features
X= dataset.iloc[:,:-1]
y= dataset.iloc[:,-1]
```

```python
X.head()
```

```python
X.shape
```

```python
y.shape
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split 
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=10)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
X_test.shape
```

```python
y_test.shape
```

```python
## Feture Scaling
# Standardize the dataset
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()
```

```python
scaler
```

```python
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
X_train
```

```python
X_test
```

# Model Training

```python
from sklearn.linear_model import LinearRegression
```

```python
regression= LinearRegression()
```

```python
regression.fit(X_train,y_train)
```

```python
# Printing the coefficients and intercepts
```

```python
print(regression.coef_)
```

```python
print(regression.intercept_)
```

```python
# Prediction for the test data
reg_pred= regression.predict(X_test)
```

```python
reg_pred
```

```python
# Assumptions of Linear Regression

#The graph is linear which means the model is good
plt.scatter(y_test,reg_pred)
plt.xlabel("Test Truth Data")
plt.ylabel("Test Predicted Data")
```

```python
# residuals
residuals= y_test-reg_pred
```

```python
# It is little bit skewed on the right side because of the outiers.
# Whenever we are trying to create a displot for the residuals, it should always give normal/gaussian distribution.
sns.displot(residuals,kind="kde")
```

```python
# Scatter plot with prediction and residuals
# If it is uniformly distributed then the model is said to be good.
# homoscedasticity
plt.scatter(reg_pred,residuals)
```

```python
# Performance Metrics
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
print(mean_squared_error(y_test,reg_pred))
print(mean_absolute_error(y_test,reg_pred))
print(np.sqrt(mean_squared_error(y_test,reg_pred)))
```

# R square and adjusted R square

```python
from sklearn.metrics import r2_score
score= r2_score(y_test,reg_pred)
print(score)
```

```python
# adjusted r square
# display adjusted r square
1 - (1-score)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
```
